```python

#removing "v1","v2" etc from end of arxiv_id in both arxiv_papers and segmented_tables
arxiv_papers['arxiv_id'] = arxiv_papers['arxiv_id'].apply(lambda x: x[:-2])
segmented_tables_annotations['arxiv_id'] = segmented_tables_annotations['arxiv_id'].apply(lambda x : x[:-2])
```

```python
arxiv_papers_seg_tables = arxiv_papers.merge(segmented_tables_annotations, how='inner',on='arxiv_id')
```

```python
arxiv_papers_seg_tables.shape
```

```python
arxiv_papers_seg_tables.head()
```

```python
#create records column
def return_records(row):
  return_records = []
  for table in row['tables_y']:
    table_records = table['records']
    if(len(table_records)>0):
      for record in table_records:
        return_records.append(record)
  return return_records
```

```python

arxiv_papers_seg_tables['records'] = arxiv_papers_seg_tables.apply(return_records,axis=1)
```

```python

arxiv_papers_seg_tables.head()
```

```python
arxiv_papers_seg_tables_with_atleast_one_record = arxiv_papers_seg_tables[arxiv_papers_seg_tables.records.map(len) > 0]
```

```python
arxiv_papers_seg_tables_with_atleast_one_record
```

```python

#expand records into each row
arxiv_papers_seg_with_records_expanded  = pd.DataFrame()
for index,row in arxiv_papers_seg_tables_with_atleast_one_record.iterrows():
  temp_df = pd.DataFrame(columns = ['arxiv_id','title','fold','record_id','record'])
  for index,record in enumerate(row['records']):
    # print(record)
    # print(record['task'])
    df = pd.Series({'arxiv_id': row['arxiv_id'], 'title': row['title'],'fold':row['fold'],'record_id':index+1,'record':record})
    # print(df)
    temp_df = temp_df.append(df,ignore_index=True)
  arxiv_papers_seg_with_records_expanded = arxiv_papers_seg_with_records_expanded.append(temp_df,ignore_index=True)
```

```python

arxiv_papers_seg_with_records_expanded_final = pd.concat([arxiv_papers_seg_with_records_expanded.drop(['record'], axis=1), arxiv_papers_seg_with_records_expanded['record'].apply(pd.Series)], axis=1)
```

```python
arxiv_papers_seg_with_records_expanded_final.head()
```

```python
arxiv_papers_seg_with_records_expanded_final['arxiv_id'] = arxiv_papers_seg_with_records_expanded_final['arxiv_id'].astype('str')
```

```python
def extract_year(row):
  return '20' + row['arxiv_id'][0:2]
```

```python
df = arxiv_papers_seg_with_records_expanded_final
```

```python
df['year'] = df.apply(lambda row: extract_year(row), axis=1)
```

```python
def extract_month(row):
  return row['arxiv_id'][2:4]
```

```python
df['month'] = df.apply(lambda row: extract_month(row), axis=1)
```
